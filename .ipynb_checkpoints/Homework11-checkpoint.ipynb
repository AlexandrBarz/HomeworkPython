```python
from sympy import *
x = Symbol("x")
```

```python
function = (sin(x)/2 + cos(x)/2) ** 2 * sin(x)
function
```

**Вычисление корней функции**

```python
func_radical = solve(function, x)
for i in func_radical:
    print(i.evalf())
```

Вычисление производной функции

```python
diff_function = diff(function)
diff_function
```

**Построение графика функции и ее производной**

```python
graf = plot(function,diff_function,(x,-2*pi,2*pi))
```

Вычисление корней производной функции

```python
diff_func_radical = solve(diff_function, x)
diff_func_radical.append(pi)
for i in diff_func_radical:
    print(i.evalf())
```

**Вычисление вершин**

```python
for x in diff_func_radical:
    y = (sin(x)/2 + cos(x)/2) ** 2 * sin(x)
    print(f'x = {round(x.evalf(),3)}, y = {round(y.evalf(),3)}')
```

**Определение промежутков, на котором f > 0**

```python
pprint(solve(function > 0, x), use_unicode=True)
```

**Определение промежутков, на котором f < 0**

```python
pprint(solve(function < 0, x), use_unicode=True)
```

**Определение интервалов, на которых функция возрастает и убывает**

```python
for i in range(len(diff_func_radical)-1):
    x = (diff_func_radical[i] + diff_func_radical[i+1])/2
    res = (-sin(x) + cos(x))*(sin(x)/2 + cos(x)/2)*sin(x) + (sin(x)/2 + cos(x)/2)**2*cos(x) 
    if res > 0:
        print('Функция возрастает на отрезке')
        print(Interval(round(diff_func_radical[i].evalf(), 3), round(diff_func_radical[i+1].evalf(), 3)))
    else:
        print('Функция убывает на отрезке')
        print(Interval(round(diff_func_radical[i].evalf(), 3), round(diff_func_radical[i+1].evalf(), )))
```
